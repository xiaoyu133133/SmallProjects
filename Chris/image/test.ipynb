残差网络

```python
class Resblock(nn.Module):
   def __init__(self, in_channels, hid_channels):
         super().__init__()
         self.in_channels = in_channels
         self.hid_channels = hid_channels
         self.cnn=nn.Sequential(
             nn.Conv2d(in_channels, hid_channels, kernel_size=3, padding=1),
             nn.BatchNorm2d(hid_channels),
             nn.SiLU(),
             nn.Conv2d(hid_channels, in_channels, kernel_size=5, padding=1), 
             nn.BatchNorm2d(in_channels),
             nn.SiLU(),
             nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1),
         )
```

```python
def forward(self, x):
        return x + self.cnn(x)
class ResNet(nn.Module):
      def __init__(self,in_channel,out_channels,num_classes,hid_channels,res_channels,num_resblocks):
        super().__init__()
        self.in_channel=in_channel
        self.num_classes=num_classes
        self.res_channels=res_channels
        self.hid_channels=hid_channels
        self.num_resblocks=num_resblocks
        self.encoder=nn.Sequential(
          

         nn.Conv2d(in_channel,res_channels,kernel_size=3,padding=1   ),
                   nn.SiLU(),
           )
        self.decoder= nn.Conv2d(res_channels,out_channels,kernel_size=3,padding=1)
        self.hid_res_layers=nn.Sequential(
                *[Resblock(res_channels,hid_channels) for _ in range(num_resblocks)]
           )
      def forward(self,x):
        y=self.encoder(x)
               
        y=self.hid_res_layers(y)
        y=self.decoder(y)
        return y.mean(dim=[-2,1])
```
